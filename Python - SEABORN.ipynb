```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# seaborn  comes with some pre loaded datasets.
tips = sns.load_dataset('tips')
```

```python
tips.head(3)
```

# PLOT TYPES:
 
Dist plot : for uni variable distribution.i.e. single variables distribution.

Joint plot : for combining two different single distribution plots.

Pairplot :  for making plots for every pair of numeric columns.

Rugplot : it works like distplot but gives dash output like a bar code.

# Dist Plot

it gives histogram.

Syntax :  

sns.distplot(column_name)

```python
sns.distplot(tips['total_bill'],kde=False,bins=30,color='g')
sns.set_context('notebook')
plt.figure(figsize=(12,2))
# the curvy line in the chart is called KDE : kernel density estimation . we can remove it.
# bins=number :  to change the number of bars in the chart.
```

# JOINT PLOT

joining two dist plots.

syntax:

sns.jointplot(x='column_name', y='column_name', data=dataframe_name)

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')   # kind='hex','reg','kde'
```

# PAIRPLOT

syntax :

sns.pairplot(dataframe_name, hue=' ')  

as it makes plots for every pair of numeric column.we can set condition of non-numeric column by mentioning in hue=''.

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')    # palette='' : to change colors.
```

# Rugplot

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

# CATEGORICAL PLOTS

BARPLOTS() : for summarizing the categorical data which by default in mean().

COUNTPLOT() : only x value is set as it only count values.

 # Barplots  
 
 Syntax:
 
 sns.barplot(x='non-numeric_col_name', y='numeric_col_name', data=df_name)

```python
# visualisation of groupby() function.
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)      # by default it shows mean().

#  estimator=aggregate_function : to return  functions like np.mean,np.std,np.sum
```

# Countplot

counts values

syntax:  sns.countplot(x='non_num_column',data=df_name)

```python
sns.countplot(x='sex',data=tips)
```

# BOXPLOT

```python
sns.boxplot(x='day',y='total_bill',data=tips)
# it shows max,min,first quartile,third quartile and median..
```

```python
# for more on boxplots
print('http://www.physics.csbsju.edu/stats/box2.html')
```

# violinplot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue="smoker",split=True)
```

# Stripplot

it is a scatter plot.syntax:

sns.stripplot(x='non_num',y='num_col',data=df_name,jitter=True)

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

# swarmplot = stripplot + violinplot

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True) # not good for big data.
```

# Factorplot now renamed as catplot

generalised plot of all categorised plots

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='box')

# kind='bar'.'strip','violin','swarm','box'
```

# MATRIXPLOT

sns.heatmap(matrix_name)

sns.clustermap(matrix_name)

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tc=tips.corr()
```

# HEATMAP

in heatmap, columns and rows should correlate. we can do this by using pivot_table to convert df into matrices. 

```python
sns.heatmap(tc,vmin=0,vmax=1,annot=True,cmap='coolwarm') # annot=True :will also display the actual values.
```

```python
fl = flights.pivot_table(index='month',values='passengers',columns='year')
fl
# df_name.pivot_table(index='col_name',columns='col_name',values='num_col_name') : used to convert dataframes 
# into matices which is first step when using sns.heatmap(matrix_name).
```

```python
sns.heatmap(fl,cmap='magma',linewidths=4,linecolor='white')  #cmap=' '=colormap
```

# clustermap

it clusters the similar values in a heirarchial form.

```python
sns.clustermap(fl,standard_scale=1)
```

# PairGrid

```python
ir=sns.load_dataset('iris')
ir.head()
```

```python
ir['species'].unique()
```

```python
# we can modify pairplot by using sns.PairGrid which will give us empty plots.

# sns.pairplot(ir)
gr=sns.PairGrid(ir)

# now we can decided the which plot we want in which grid
gr.map_diag(sns.distplot)
gr.map_upper(plt.scatter)
gr.map_lower(sns.kdeplot)
```

# FacetGrid

```python
fg = sns.FacetGrid(tips,row='time',col='smoker')

fg.map(sns.distplot,'total_bill')
```

```python
# more than one argument in facetgrid.
fg = sns.FacetGrid(tips,row='time',col='smoker')

fg.map(plt.scatter,'total_bill','tip')
```

# REGRESSION PLOTS

```python
# LMPLOT(x=num_col , y=num_col ,data=df_name , hue=optional)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
           scatter_kws={'s':40})  # scatter_kws={'s':num}  : to change the size of markers
```

```python
# to make two plots by seperating on the baseis of categorical column.
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,height=8)

# we will get different plots.
# inplace of hue= , we can use this col= and row=  option.
# size=  has been renamed to height=   
# we can still add hue = col_name  argument

# to change aspect ratio and size
# aspect=0.6, it is the ratio between height and width.
```

# Styling plots

```python
# sns.set_style() : to set plot style for all the plots to be made after this. syntax:

sns.set_style('whitegrid')
sns.countplot('sex',data=tips)

# to remaove spines : spines are the top and right lines of the plot.
sns.despine(left=True)
```
