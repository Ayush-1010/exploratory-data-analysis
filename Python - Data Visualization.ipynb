```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('df1',index_col='Unnamed: 0')
# to mark a column as an index in read_csv use :  index_col='column_name'
```

```python
df1.index
```

```python
# we can make histograms or other plots on the basis of single columns of  dataframe.
# method 1
sns.distplot(df1['A'],kde=False)
```

```python
# method 2
df1['A'].plot(kind='hist',bins=30)

# or
# df1['A'].plot.hist()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
# area plot

df2.plot.area(alpha=0.5)
```

```python
df2.plot.bar(stacked=True)
```

```python
# LINE Graph
df1.plot.line(x='C',y='A')
```

```python
df1.plot.scatter(x="A",y="B",s=df1["C"]*10) # s for size and adding a third comparison.
```

```python
# box plot in df

df2.plot.box()
# it takes every column on x axis.
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['A','B'])
```

```python
df.head()
```

```python
# hexplot
df.plot.hexbin(x='A',y='B',gridsize=20,cmap='coolwarm')
```

```python
df.plot.kde()
```
